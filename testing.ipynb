```python
import os


def print_folder_structure(startpath):
    for root, dirs, files in os.walk(startpath):
        # Filter out hidden directories and files
        dirs[:] = [d for d in dirs if not d.startswith(".")]
        files = [f for f in files if not f.startswith(".")]

        level = root.replace(startpath, "").count(os.sep)
        indent = " " * 4 * level
        print(f"{indent}{os.path.basename(root)}/")
        subindent = " " * 4 * (level + 1)
        for file in files:
            print(f"{subindent}{file}")
```

```python
# Get the current working directory
current_folder = os.getcwd()

# Call the function to print the folder structure
print_folder_structure(current_folder)
```

```python
import os
from typing import Union


def count_png_files(folder_path: str) -> int:
    """
    Counts the number of .png files in a nested directory structure.

    Parameters:
        folder_path (str): The path to the root folder.

    Returns:
        int: The total number of .png files found within the folder and its subdirectories.
    """
    count = 0
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            # Check file extension in a case-insensitive manner
            if file.lower().endswith(".png"):
                count += 1
    return count
```
